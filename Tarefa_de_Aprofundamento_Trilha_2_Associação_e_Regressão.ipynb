# **TRATANDO OS DADOS:**
Para tratar os dados, vamos separar os dados das colunas em 4 grupos, Enormes(gg), grandes(g), medios(m) e pequenos(p). Isso vai fazer o apriori achar padrões mais facilmente.

```python
#Separando os dados em grupos
sepal_length = pd.qcut(iris[0], 4, labels=['sepal_length_gg','sepal_length_g', 'sepal_length_m','sepal_length_p'])
sepal_width = pd.qcut(iris[1], 4, labels=['sepal_width_gg','sepal_width_g', 'sepal_width_m','sepal_width_p'])
petal_length = pd.qcut(iris[2], 4, labels=['petal_length_gg','petal_length_g', 'petal_length_m','petal_length_p'])
petal_width = pd.qcut(iris[3], 4, labels=['petal_width_gg','petal_width_g', 'petal_width_m','petal_width_p'])

df_disc = pd.concat([sepal_length,sepal_width,petal_length,petal_width,iris['Iris-setosa']],axis=1)
df_disc.head()
```

```python
#Fazendo a discretização dos dados em verdades e falsos
df_disc = df_disc.values.tolist()

encoder = TransactionEncoder()
iris_array = encoder.fit_transform(df_disc)
df = pd.DataFrame(iris_array, columns=encoder.columns_)
df
```

```python
Rules = list(apriori(df_disc,  min_length=4, min_support=0.2, min_confidence=0.8))

exibindo(Rules)
```
